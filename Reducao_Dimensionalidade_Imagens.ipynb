```python
#importar as bibliotecas essenciais para o processamento de imagens
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = Image.open('lena.jpg')
imagem_array = np.array(imagem)

# Função para converter para tons de cinza
def para_tons_de_cinza(imagem):
    return np.dot(imagem[...,:3], [0.2989, 0.5870, 0.1140])

# Função para binarizar
def binarizar(imagem, limiar=127):
    return (imagem > limiar) * 255

# Converter para tons de cinza
imagem_cinza = para_tons_de_cinza(imagem_array)

# Binarizar a imagem
imagem_binaria = binarizar(imagem_cinza)

# Exibir as imagens
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.imshow(imagem_array)
plt.title('Original')
plt.axis('off')

#Exibir a imagem convertida para tons de cinza
plt.subplot(132)
plt.imshow(imagem_cinza, cmap='gray')
plt.title('Tons de Cinza')
plt.axis('off')

#Exibir a imagem binarizada (convertida para preto e branco)
plt.subplot(133)
plt.imshow(imagem_binaria, cmap='gray')
plt.title('Binarizada')
plt.axis('off')

# Exibir todas as imagens processadas
plt.show()

# Salvar as imagens processadas
Image.fromarray(imagem_cinza.astype(np.uint8)).save('lena_cinza.jpg')
Image.fromarray(imagem_binaria.astype(np.uint8)).save('lena_binaria.jpg')
```
